```python
#Bibliotecas utilizadas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import os
```

```python
def extrair_senioridade(texto):
    """
    Extrai o nível de senioridade do texto da vaga.
    Parâmetros:
        texto (str): texto contendo título e descrição da vaga.
    Retorna:
        str: nível de senioridade categorizado.
    """
    if re.search(r'\bsênior\b|\bsenior\b', texto, re.IGNORECASE):
        return "Sênior"
    elif re.search(r'\bpleno\b', texto, re.IGNORECASE):
        return "Pleno"
    elif re.search(r'\bjúnior\b|\bjunior\b', texto, re.IGNORECASE):
        return "Júnior"
    elif re.search(r'\bestagiário\b|\bestagiario\b', texto, re.IGNORECASE):
        return "Estagiário"
    else:
        return "Não especificado"
```

```python
def extrair_idioma(texto):
    """
    Extrai o idioma requerido na vaga a partir do texto.
    """
    if re.search(r'\binglês\b|\bingles\b', texto, re.IGNORECASE):
        return "Inglês"
    elif re.search(r'\bespanhol\b', texto, re.IGNORECASE):
        return "Espanhol"
    else:
        return "Não especificado"
```

```python
def extrair_certificado(texto):
    """
    Extrai certificações relacionadas à nuvem a partir do texto.
    """
    if re.search(r'\bazure\b', texto, re.IGNORECASE):
        return "Microsoft Azure"
    elif re.search(r'\baws\b', texto, re.IGNORECASE):
        return "Amazon Web Service"
    elif re.search(r'\bgoogle\b|\bgcp\b', texto, re.IGNORECASE):
        return "Google Cloud"
    else:
        return "Não especificado"
```

```python
def criar_features(df):
    """
    Aplica as funções de extração e adiciona colunas ao DataFrame.
    """
    df["title"] = df["title"].fillna("").str.strip()
    df["description"] = df["description"].fillna("").str.strip()
    df["titulo_descr"] = df["title"] + " " + df["description"]

    df["senioridade"] = df["titulo_descr"].apply(extrair_senioridade)
    df["idioma"] = df["description"].apply(extrair_idioma)
    df["certificado"] = df["description"].apply(extrair_certificado)
    return df
```

```python
def plotar_contagem(df, coluna, titulo, paleta="Blues_d"):
    """
    Cria gráfico de barras da contagem dos valores da coluna.
    """
    sns.set(style="whitegrid")
    plt.figure(figsize=(8, 5))
    order = df[coluna].value_counts().index
    ax = sns.countplot(x=coluna, data=df, order=order, palette=paleta)
    plt.title(titulo, fontsize=14)
    plt.xlabel(coluna.capitalize())
    plt.ylabel("Número de Vagas")
    
    # Exibir valores em cima das barras
    for i, count in enumerate(df[coluna].value_counts()):
        ax.text(i, count + 0.5, str(count), ha='center', va='bottom', fontsize=10)
    
    plt.tight_layout()
    plt.show()
```

```python
file_path = os.path.join("..", "data", "raw", "linkedin_datascience_jobs.csv")
df = pd.read_csv(file_path, encoding="utf-8-sig")
df.head()
```

```python
# Criar novas features
df = criar_features(df)
```

```python
print("Distribuição de Senioridade:")
print(df["senioridade"].value_counts())
plotar_contagem(df, "senioridade", "Distribuição da Senioridade nas Vagas")
```

```python
print("\nDistribuição de Idiomas:")
print(df["idioma"].value_counts())
plotar_contagem(df, "idioma", "Distribuição dos Idiomas nas Vagas", paleta="coolwarm")
```

```python
print("\nDistribuição de Certificados:")
print(df["certificado"].value_counts())
plotar_contagem(df, "certificado", "Distribuição dos Certificados de Nuvem nas Vagas", paleta="Set3")
```

```python
output_path = os.path.join("..", "data", "processed", "linkedin_datascience_jobs_fe.csv")
df.to_csv(output_path, index=False, encoding="utf-8-sig")
print(f"\nArquivo com features salvos em {output_path}")
```
